```python
%matplotlib inline

import sys
import os
import numpy as np
import pandas as pd
import glob
import h5py
import time

import matplotlib.pyplot as plt
import matplotlib as mpl
from tomo_encoders import viewer, DataFile, Patches
```

```python
def get_ds(filename):
    data_path = '/data02/MyArchive/AM_part_Xuan'
    ct_fpath = os.path.join(data_path, 'data', filename + '_rec_1x1_uint16.hdf5')
    return DataFile(ct_fpath, data_tag = "data", tiff = False, VERBOSITY = 0)
```

```python
ds_1 = get_ds('mli_L206_HT_650_L3')
ds_2 = get_ds('mli_L206_HT_650_L4')
```

```python
img_1 = ds_1.read_slice(axis = 1, slice_idx=ds_1.d_shape[1]//2)
img_2 = ds_2.read_slice(axis = 1, slice_idx=ds_2.d_shape[1]//2)
```

```python
fig, ax = plt.subplots(2,1, figsize = (8,6))
ax[0].imshow(img_1, cmap = 'gray')
ax[1].imshow(img_2, cmap = 'gray')
fig.tight_layout()
```
